```python
# Importando as bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import os
import kagglehub
```

```python
# Carregar a base de dados

path = kagglehub.dataset_download("ankushpanday1/india-statewise-daily-temperature-for-dec-2024")
os.environ['KAGLE_CONFIG_DIR'] = '/content'
print("Path to dataset files:", path)
os.listdir(path)
```

```python
path = '/root/.cache/kagglehub/datasets/ankushpanday1/india-statewise-daily-temperature-for-dec-2024/versions/1'
df = pd.read_csv(path + '/India_December_2024_Temperature.csv')
```

```python
# Verificando todas as colunas com os tipos de dados e a quantidade de conteúdos nulos
df.info()
```

```python
# Redundante, sabemos da linha anterior que não temos linha nula
df.isnull().sum()
```

```python
# Definir os dias de dezembro (eixo X)
dias_dezembro = [f"Dec {i}" for i in range(1, 32)]

# Número de cidades (linhas) no dataset
num_cidades = len(df)

# Criar uma figura com subgráficos
fig, axes = plt.subplots(num_cidades, 1, figsize=(10, 2 * num_cidades))  # Uma linha por cidade

# Iterando sobre cada linha (cidade) e plotando nos subgráficos
for i, (index, row) in enumerate(df.iterrows()):
    axes[i].plot(dias_dezembro, row[1:], label=row['State'])
    axes[i].grid(True, axis='both', linestyle='--', color='gray', alpha=0.7)
    axes[i].set_title(row['State'])
    axes[i].set_xlabel('Dia de Dezembro')
    axes[i].set_ylabel('Temperatura (°C)')
    axes[i].tick_params(axis='x', rotation=45)

# Ajuste do layout para não sobrepor os subgráficos
plt.tight_layout()
plt.show()
```
