# ML:

```python
# Vamos usar para conveter a variável categorica em números

from sklearn.preprocessing import OneHotEncoder
```

```python
# Passando toda a coluna species

ohe_df = pd.DataFrame(df['species'])
```

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
# Criams um novo dataframe com as alterações para classificar cada uma das especies diferentes
# Vou usar isso para verificar se existe alguma diferença na mudança da eficiência

encoded_species = encoder.fit_transform(df[['species']])
encoded_df = pd.DataFrame(encoded_species, columns=encoder.get_feature_names_out(['species']))
```

```python
# Mostrando que o dataset não foi alterado mas o novo contém as novas colunas
df
```

```python
encoded_df
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

X = df.drop(columns=['species'])
y = df['species']

X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.20,random_state=42)
```

# Regressão Logística:

```python
from sklearn.linear_model import LogisticRegression

#Criando e treinando o modelo
model = LogisticRegression()
model.fit(X_train, y_train)

# Faça previsões
y_pred = model.predict(X_test)

# Avalie o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia: {accuracy}')
```

# K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

#Criando e treinando o modelo
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

# Faça previsões
y_pred = model.predict(X_test)

# Avalie o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia: {accuracy}')
```

```python
# segundo modelo

X = df.drop(columns=['species'])
y = encoded_df

X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.20,random_state=42)

#Criando e treinando o modelo
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

# Faça previsões
y_pred = model.predict(X_test)

# Avalie o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia: {accuracy}')
```

# Árvores de Decisão:

```python
from sklearn.tree import DecisionTreeClassifier

#Criando e treinando o modelo
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Faça previsões
y_pred = model.predict(X_test)

# Avalie o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia: {accuracy}')
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

#Criando e treinando o modelo
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Faça previsões
y_pred = model.predict(X_test)

# Avalie o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia: {accuracy}')
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

#Criando e treinando o modelo
model = GaussianNB()
model.fit(X_train, y_train)

# Faça previsões
y_pred = model.predict(X_test)

# Avalie o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia: {accuracy}')
```

# Rede Neural Artificial

```python
from sklearn.neural_network import MLPClassifier

#Criando e treinando o modelo
model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000)
model.fit(X_train, y_train)

# Faça previsões
y_pred = model.predict(X_test)

# Avalie o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia: {accuracy}')
```
